```python
list_1=[]
for i in data['rawData']:
    dict_1={}
    dict_1['Country Name']=i['Country_Region']
    dict_1['Confirmed Values']=i['Confirmed']
    dict_1['Deaths']=i['Deaths']
    list_1.append(dict_1)
list_1
```

```python
field_name= ['Country Name', 'Confirmed Values', 'Deaths']
with open('Covid_Version_1.csv', mode='w', newline='') as csv_file:
    push = csv.DictWriter(csv_file,fieldnames=field_name)
    push.writeheader()
    push.writerows(list_1)
```

```python
with open('output.csv', mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Country Name', 'Confirmed Values', 'Deaths'])
    for i in data['rawData']: 
        writer.writerow([i['Country_Region'], i['Confirmed'], i['Deaths']])
```
